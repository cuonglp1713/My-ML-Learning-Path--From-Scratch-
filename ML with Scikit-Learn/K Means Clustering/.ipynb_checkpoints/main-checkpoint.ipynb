## K Means Clustering model 

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=2)
model.fit(data.drop('Private', axis=1))
```

```python
model.cluster_centers_
```

## Predictions & Evaluations 

```python
def convert(private):
    if private == 'Yes':
        return 1
    else:
        return 0
```

```python
data['Cluster'] = data['Private'].apply(convert)

data.head()
```

```python
y_predict = model.labels_
y_predict
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(data['Cluster'], y_predict))
print(confusion_matrix(data['Cluster'], y_predict))
```
